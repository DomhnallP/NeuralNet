```python
# this file was created by following youtube user sentdex's tutorial, fullcredit to him!
# link here: https://www.youtube.com/watch?v=wQ8BIBpya2k
import tensorflow as tf
import keras
mnist = keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data() 

x_train = keras.utils.normalize(x_train, axis=1)
x_test = keras.utils.normalize(x_test, axis=1)

model = keras.models.Sequential()
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(128,activation=tf.nn.relu))
model.add(keras.layers.Dense(128,activation=tf.nn.relu))
model.add(keras.layers.Dense(10, activation=tf.nn.softmax))

model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

model.fit(x_train, y_train, epochs=3)
```

```python
val_loss, vall_acc = model.evaluate(x_test,y_test)
print(val_loss, vall_acc)
```

```python
model.save("num_interpreter.model")
```

```python
new_model = keras.models.load_model('num_interpreter.model')
```

```python
predictions = new_model.predict([x_test])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

test_selection = random.sample(range(0, len(predictions)), 20)
print(test_selection)
fig=plt.figure(figsize=(20,20))
columns = 5
rows = (len(test_selection)/columns)+1
for image,i in zip(test_selection, range(0,len(test_selection))):
    ax = fig.add_subplot(rows, columns, i+1)
    fig.tight_layout()
    ax.set_title("Neural Net Guess: " + str(np.argmax(predictions[image])), y=1.08)
    plt.imshow(x_test[image])
    
```
